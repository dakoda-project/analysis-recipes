# Advanced Metadata

```python
from dakoda.corpus import DakodaCorpus, DakodaPublicCorpusName, MetaData

merlin = DakodaCorpus(DakodaPublicCorpusName.MERL_L2)
doc = merlin.random_doc()
meta = doc.meta

for meta_var in meta.iter_flat():
    print(meta_var)
```

```python
meta.print_schema_tree()
```

```python
for lang in meta.learner.language:
    status = lang.learner_language_status # LangStatus (e.g., L1, L2, etc.)
    lang_code = lang.learner_language_iso639_3 # Lingua object
    if "L1" in str(status):
        print(f"L1 Language: {lang_code.iso_code_639_3}")
```
