# Text Visualization

```python
from dakoda.corpus import DakodaCorpus, DakodaPublicCorpusName
from dakoda.uima import load_dakoda_typesystem
from cas_visualizer.visualizer import SpanVisualizer
from IPython.display import display, HTML
```

```python
wtld = DakodaCorpus(DakodaPublicCorpusName.MERL_L2)
doc = wtld.random_doc()

cas = doc.learner._view 
vis = SpanVisualizer(load_dakoda_typesystem())
vis.selected_span_type = SpanVisualizer.HIGHLIGHT
vis.add_type('Sentence', label = 'S', color="#DDDDDD")

html = vis.visualize(cas)
display(HTML(html))
```

```python
doc = wtld.random_doc()
cas = doc.learner._view 
vis_lemma = SpanVisualizer(load_dakoda_typesystem())
vis_lemma.selected_span_type = SpanVisualizer.UNDERLINE
vis_lemma.add_type('de.tudarmstadt.ukp.dkpro.core.api.segmentation.type.Lemma', feature = 'value', color="#FFA500")

html = vis_lemma.visualize(cas)
display(HTML(html))
```

```python
doc = wtld.random_doc()
cas = doc.learner._view 
vis_lemma = SpanVisualizer(load_dakoda_typesystem())
vis_lemma.selected_span_type = SpanVisualizer.HIGHLIGHT
# full type name needed due to ambiguity with another 'POS' type
vis_lemma.add_type(
    name = 'de.tudarmstadt.ukp.dkpro.core.api.lexmorph.type.pos.POS',
    feature = 'PosValue'
)

html = vis_lemma.visualize(cas)
display(HTML(html))
```
