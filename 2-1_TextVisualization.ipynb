# Text Visualization

```python
from dakoda.corpus import DakodaCorpus
from dakoda.uima import load_dakoda_typesystem
from cas_visualizer.visualizer import SpanVisualizer
from IPython.display import display, HTML
```

```python
wtld = DakodaCorpus("data/swiko_wtld/DaF")
doc = wtld.random_doc()

cas = doc.learner._view 
vis = SpanVisualizer(load_dakoda_typesystem())
vis.selected_span_type = SpanVisualizer.HIGHLIGHT
vis.add_type('Sentence', default_label = 'S', color="#DDDDDD")

html = vis.visualize(cas)
display(HTML(html))
```

```python
doc = wtld.random_doc()
cas = doc.learner._view 
vis_lemma = SpanVisualizer(load_dakoda_typesystem())
vis_lemma.selected_span_type = SpanVisualizer.UNDERLINE
vis_lemma.add_type(type_name = 'Lemma', feature_name = 'value', color="#FFA500")

html = vis_lemma.visualize(cas)
display(HTML(html))
```

```python
doc = wtld.random_doc()
cas = doc.learner._view 
vis_lemma = SpanVisualizer(load_dakoda_typesystem())
vis_lemma.selected_span_type = SpanVisualizer.HIGHLIGHT
# full type name needed due to ambiguity with another 'POS' type
vis_lemma.add_type(type_name = 'de.tudarmstadt.ukp.dkpro.core.api.lexmorph.type.pos.POS', feature_name = 'PosValue')

html = vis_lemma.visualize(cas)
display(HTML(html))
```

```python
from spacy.displacy.render import DependencyRenderer 
from recipe_util import dep_to_dict

doc = wtld.random_doc()
sent = doc.learner.sentences[0]

struct = dep_to_dict(cas=doc.cas, span=sent.span, view_name='ctok')

# words = [
#     {'text': 'Hallo', 'tag': 'ITJ'},
#     {'text': ',', 'tag': '$,'},
#     {'text': 'ich', 'tag': 'PPER'},
#     {'text': 'heisse', 'tag': 'VVFIN'},
#     {'text': 'Maria', 'tag': 'NE'}
# ]

# arcs = [
#     {'start': 0, 'end': 1, 'label': 'punct', 'dir': 'left'},
#     {'start': 2, 'end': 4, 'label': 'nsubj', 'dir': 'left'},
#     {'start': 3, 'end': 4, 'label': 'cop', 'dir': 'left'},
#     {'start': 0, 'end': 4, 'label': 'parataxis', 'dir': 'left'}
# ]

renderer = DependencyRenderer(options={
    'compact': True, 
    'bg': '#f8f8f8', 
    'color': 'black', 
    'font': 'Arial'}
)
html = renderer.render([{"words": struct['words'], "arcs": struct['arcs']}])
# html = renderer.render([{"words": words, "arcs": arcs}])

display(HTML(html))
```

```python
renderer = DependencyRenderer()
html = renderer.render([{"words": struct['words'], "arcs": struct['arcs']}])
# html = renderer.render([{"words": words, "arcs": arcs}])
svg = renderer.render_svg(render_id=1, words=struct['words'], arcs = struct['arcs'])

svg_html = f"""
<div style="width:400px; height:200px; overflow:auto;">
    <svg style="transform: scale(0.5); transform-origin: top left;">
        {svg}
    </svg>
</div>
"""
display(HTML(svg_html))
```
